## –ó–∞–¥–∞–Ω–∏–µ 1

–ü–æ–¥–≥–æ—Ç–æ–≤—å –¥–∞–Ω–Ω—ã–µ –¥–ª—è –æ—Ç–≤–µ—Ç–∞ –Ω–∞ –≤–æ–ø—Ä–æ—Å: ¬´__–ö–∞–∫ –∏–∑–º–µ–Ω—è–ª–∞—Å—å –µ–∂–µ–º–µ—Å—è—á–Ω–∞—è –≤—ã—Ä—É—á–∫–∞ –ø–æ –∫–∞–∂–¥–æ–º—É —Ç–∏–ø—É —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ–∫ –≤ —Ç–µ—á–µ–Ω–∏–µ 2020 –≥–æ–¥–∞?__¬ª. 

1. –ò–∑—É—á–∏ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º `payments` –∏ –¥–ª—è –Ω–∞—á–∞–ª–∞ –ø–µ—Ä–µ–≤–µ–¥–∏ –¥–∞—Ç—ã –≤ —Ñ–æ—Ä–º–∞—Ç `datetime`.

2. –° –ø–æ–º–æ—â—å—é —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–∏ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –≤—ã–±–µ—Ä–∏ –ø–ª–∞—Ç–µ–∂–∏ —Ç–æ–ª—å–∫–æ –∑–∞ 2020 –≥–æ–¥.

3. –û–±—Ä–∞—Ç–∏ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –ø–ª–∞—Ç–µ–∂–∏ —Ö—Ä–∞–Ω—è—Ç—Å—è –≤ —Ä–∞–∑–Ω—ã—Ö –≤–∞–ª—é—Ç–∞—Ö ‚Äî –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞ –ø—Ä–∏–≤–µ–¥–µ–º –≤—Å–µ –∫ —Ä—É–±–ª—è–º. –°–Ω–∞—á–∞–ª–∞ –¥–æ–±–∞–≤—å –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å—Ç–æ–ª–±–µ—Ü `fx_rate` —Å –∫—É—Ä—Å–æ–º –≤–∞–ª—é—Ç—ã –∏ –∑–∞–ø–æ–ª–Ω–∏ –µ–≥–æ —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ `apply()`: –µ—Å–ª–∏ —ç—Ç–æ USD, —Ç–æ –±–µ—Ä–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π `USD_RATE`, –µ—Å–ª–∏ RUR ‚Äî —Å—Ç–∞–≤–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ `1`. –ü—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ –Ω—É–∂–Ω–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è —Ä–µ–∞–ª–∏–∑—É–µ—Ç —ç—Ç—É –Ω–µ—Å–ª–æ–∂–Ω—É—é –ª–æ–≥–∏–∫—É.

4. –î–æ–±–∞–≤—å –µ—â–µ –æ–¥–∏–Ω —Å—Ç–æ–ª–±–µ—Ü `amount_rub` —Å —Å—É–º–º–æ–π –ø–ª–∞—Ç–µ–∂–∞ –≤ —Ä—É–±–ª—è—Ö. –î–ª—è —ç—Ç–æ–≥–æ —É–º–Ω–æ–∂—å —Å—Ç–æ–ª–±–µ—Ü —Å —Å—É–º–º–æ–π –ø–ª–∞—Ç–µ–∂–∞ –Ω–∞ —Å—Ç–æ–ª–±–µ—Ü `fx_rate` —Å –∫—É—Ä—Å–æ–º –≤–∞–ª—é—Ç.

5. –ü–æ–ª—É—á–∏ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å —Ä–∞–∑–±–∏–≤–∫–æ–π –≤—ã—Ä—É—á–∫–∏ –ø–æ –º–µ—Å—è—Ü–∞–º –∏ –ø–æ —Ç–∏–ø–∞–º —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ–∫ —Å –ø–æ–º–æ—â—å—é –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∏ —Å —Å—É–º–º–æ–π –º–µ—Ç–æ–¥–æ–º `groupby()`. 

6. –í—ã–≤–µ–¥–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ —ç–∫—Ä–∞–Ω.

ü§î *–ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏, –æ—Å–Ω–æ–≤—ã–≤–∞—è—Å—å –Ω–∞ –ø—Ä–æ–¥–µ–ª–∞–Ω–Ω–æ–º –∞–Ω–∞–ª–∏–∑–µ.* –ë–ª–∞–≥–æ–¥–∞—Ä—è –∫–∞–∫–∏–º —Ç—Ä–µ–Ω–∏—Ä–æ–≤–∫–∞–º –∫–æ–º–ø–∞–Ω–∏—è –º–æ–∂–µ—Ç –ø–æ–ª—É—á–∞—Ç—å –±–æ–ª—å—à–µ –≤—ã—Ä—É—á–∫–∏? –£ —ç—Ç–æ–≥–æ –∑–∞–¥–∞–Ω–∏—è –Ω–µ—Ç –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –æ—Ç–≤–µ—Ç–∞, –º—ã –ª–∏—à—å –ø—Ä–µ–¥–ª–∞–≥–∞–µ–º —Ç–µ–±–µ –ø–æ—Ä–∞–∑–º—ã—à–ª—è—Ç—å.

```python
payments.head()  # –ø–æ—Å–º–æ—Ç—Ä—é –¥–∞—Ç–∞—Ñ—Ä–µ–π–º payments
```

```python
payments['payment_date'] = pd.to_datetime(payments['payment_date']) # –ø–µ—Ä–µ–≤–µ–¥—É –¥–∞—Ç—ã –≤ —Ñ–æ—Ä–º–∞—Ç datetime
payments.head()  # –ø–æ—Å–º–æ—Ç—Ä—é –¥–∞—Ç–∞—Ñ—Ä–µ–π–º payments
```

```python
payments.info() # –ø–æ—Å–º–æ—Ç—Ä—é —Ñ–æ—Ä–º–∞—Ç—ã –¥–∞–Ω–Ω—ã—Ö
```

```python
# –æ—Ç—Ñ–∏–ª—å—Ç—Ä—É—é –ø–ª–∞—Ç–µ–∂–∏ –∑–∞ 2020–≥–æ–¥
payments.loc[(payments['payment_date'] > '2020-01-01') & (payments['payment_date'] < '2021-01-01')]
```

```python
# –∏–ª–∏ –¥—Ä—É–≥–∏–º —Å–ø–æ—Å–æ–±–æ–º, –∏—Å–ø–æ–ª—å–∑—É—è –º–µ—Ç–æ–¥ .dt –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ Pandas
payments.loc[payments['payment_date'].dt.year == 2020]
```

```python
# –ø—Ä–∏–≤–µ–¥—É –ø–ª–∞—Ç–µ–∂–∏ –∫ —Ä—É–±–ª—è–º, –¥–æ–±–∞–≤–∏–≤ —Å—Ç–æ–ª–±–µ—Ü fx_rate —Å –∫—É—Ä—Å–æ–º –≤–∞–ª—é—Ç—ã

USD_RATE = 73.14 # –∫—É—Ä—Å –¥–æ–ª–ª–∞—Ä–∞

def map_currency(currency): # –æ–ø—Ä–µ–¥–µ–ª—é —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ fx_rate –º–µ—Ç–æ–¥–æ–º .apply()
    if currency == 'USD':
        return USD_RATE
    elif currency == 'RUR':
        return 1


payments['fx_rate'] = payments['currency'].apply(map_currency) # —Å–æ–∑–¥–∞–¥ —Å—Ç–æ–ª–±–µ—Ü fx_rate –ø–æ —É—Å–ª–æ–≤–∏—é
```

```python
payments.describe() # –ø–æ—Å–º–æ—Ç—Ä—é, —á—Ç–æ –ø–æ–ª—É—á–∏–ª–æ—Å—å ))
```

```python
# –¥–æ–±–∞–≤–ª—é –µ—â–µ –æ–¥–∏–Ω —Å—Ç–æ–ª–±–µ—Ü amount_rub —Å —Å—É–º–º–æ–π –ø–ª–∞—Ç–µ–∂–∞ –≤ —Ä—É–±–ª—è—Ö

payments['amount_rub'] = payments['amount'] * payments['fx_rate']
```

```python
payments.head() # –ø–æ—Å–º–æ—Ç—Ä—é, —á—Ç–æ –ø–æ–ª—É—á–∏–ª–æ—Å—å
```

```python
# –ü–æ–ª—É—á—É –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å —Ä–∞–∑–±–∏–≤–∫–æ–π –≤—ã—Ä—É—á–∫–∏ –ø–æ –º–µ—Å—è—Ü–∞–º –∏ –ø–æ —Ç–∏–ø–∞–º —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ–∫

payments.groupby([payments['payment_date'].dt.month, 'workout_type'])['amount_rub'].sum() # –Ω–µ –∑–∞–±—ã–≤–∞–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å–ø–∏—Å–æ–∫
```

1. –ò–∑—É—á–∏ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º `workouts`.

3. C –ø–æ–º–æ—â—å—é `.merge()` –æ–±—ä–µ–¥–∏–Ω–∏ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—ã `user_regions` –∏ `user_payments`

5. –°–≥—Ä—É–ø–ø–∏—Ä—É–π –¥–∞–Ω–Ω—ã–µ –æ –≤—ã—Ä—É—á–∫–µ –ø–æ —Ä–µ–≥–∏–æ–Ω—É, –∞ –∑–∞—Ç–µ–º –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–π –∏—Ö.
6. –í—ã–≤–µ–¥–∏ –Ω–∞ —ç–∫—Ä–∞–Ω —Ç–æ–ø-5 —Ä–µ–≥–∏–æ–Ω–æ–≤ —Å —Å–∞–º–æ–π –º–∞–ª–µ–Ω—å–∫–æ–π –≤—ã—Ä—É—á–∫–æ–π. 
7. –í—ã–≤–µ–¥–∏ –Ω–∞ —ç–∫—Ä–∞–Ω —Ç–æ–ø-5 —Ä–µ–≥–∏–æ–Ω–æ–≤ —Å —Å–∞–º–æ–π –±–æ–ª—å—à–æ–π –≤—ã—Ä—É—á–∫–æ–π. 
8. –°–æ—Ö—Ä–∞–Ω–∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø–æ —É–±—ã–≤–∞–Ω–∏—é —Å–ø–∏—Å–æ–∫ —Ä–µ–≥–∏–æ–Ω–æ–≤ —Å –≤—ã—Ä—É—á–∫–æ–π –≤ —Ñ–∞–π–ª `task_4.xlsx` —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ `to_excel()`.

ü§î *–ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏, –æ—Å–Ω–æ–≤—ã–≤–∞—è—Å—å –Ω–∞ –ø—Ä–æ–¥–µ–ª–∞–Ω–Ω–æ–º –∞–Ω–∞–ª–∏–∑–µ.* –ö–∞–∫ —Å–µ–π—á–∞—Å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –≤—ã—Ä—É—á–∫–∞ –º–µ–∂–¥—É —Ä–µ–≥–∏–æ–Ω–∞–º–∏? –£ —ç—Ç–æ–≥–æ –∑–∞–¥–∞–Ω–∏—è –Ω–µ—Ç –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –æ—Ç–≤–µ—Ç–∞, –º—ã –ª–∏—à—å –ø—Ä–µ–¥–ª–∞–≥–∞–µ–º —Ç–µ–±–µ –ø–æ—Ä–∞–∑–º—ã—à–ª—è—Ç—å.



```python
# —Å–æ–∑–¥–∞–º –Ω–æ–≤—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º user_regions –∏–∑ –∫–æ–ª–æ–Ω–æ–∫ user_id –∏ region –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ users
user_regions = users[['user_id', 'region']] 
```

```python
user_regions.head() # –ø–æ—Å–º–æ—Ç—Ä—é, —á—Ç–æ –ø–æ–ª—É—á–∏–ª–æ—Å—å
```

```python
# c–æ–∑–¥–∞–º –Ω–æ–≤—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º user_payments –∏–∑ –∫–æ–ª–æ–Ω–æ–∫ user_id –∏ amount_rub –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ payments
user_payments = payments[['user_id', 'amount_rub']]
```

```python
user_payments.head() # –ø–æ—Å–º–æ—Ç—Ä—é, —á—Ç–æ –ø–æ–ª—É—á–∏–ª–æ—Å—å
```

```python
# –æ–±—ä–µ–¥–∏–Ω—é –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—ã user_regions –∏ user_payments –∏ –æ—á–∏—â—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ—Ç —Å—Ç—Ä–æ–∫ —Å NaN –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ (–ø—É—Å—Ç—ã–º–∏)
regional_revenue = pd.merge( user_regions, user_payments).dropna()
regional_revenue
```

<div class="alert alert-block alert-info">
    
–ú–µ—Ç–æ–¥ —Ä–µ—à–µ–Ω–∏—è –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ—á–∫—É

```python
# –î–ª—è –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –≤—ã–≤–æ–¥–∞ –≤ .groupby –≤–µ—Ä–Ω–∏—Ç–µ –æ–±—ä–µ–∫—Ç —Å –≥—Ä—É–ø–ø–æ–≤—ã–º–∏ –º–µ—Ç–∫–∞–º–∏ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∏–Ω–¥–µ–∫—Å–∞.
regional_revenue \
   .groupby('region', as_index=False) \
   .aggregate({'amount_rub':'sum'}) \
   .sort_values('amount_rub', ascending = True) # —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞
```

<div class="alert alert-block alert-info">
–ü–æ—á—Ç–∏ —Ç–æ–∂–µ —Å–∞–º–æ–µ, —Ç–æ–ª—å–∫–æ –±–µ–∑ –∏–Ω–¥–µ–∫—Å–∞
</div>

```python
regional_revenue.groupby('region').agg({'amount_rub':'sum'}).sort_values('amount_rub', ascending = True)
```

```python
# c–≥—Ä—É–ø–ø–∏—Ä—É—é –¥–∞–Ω–Ω—ã–µ –æ –≤—ã—Ä—É—á–∫–µ –ø–æ —Ä–µ–≥–∏–æ–Ω—É
regional_revenue_group = regional_revenue.groupby('region')['amount_rub'].sum()
regional_revenue_group
```

```python
regional_revenue_group.sort_values(by='amount_rub').head() # –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É—é –∏ –≤—ã–≤–µ–¥—É —Ç–æ–ø-5 —Å —Å–∞–º–æ–π –º–∞–ª–æ–π –≤—ã—Ä—É—á–∫–æ–π
```
